Создать представление с полями:

Название колонки | Источник иформации
--- | ---

Название продукта | столбец productName таблицы products

Отобразить результат вызова представления

Из полученного представления найти заказчиков и принесенную прибыль, (отсортировать по убыванию суммарной прибыли по всем), отобразить.
Учитывайте, что заказ должен быть доставлен! Shipped = True

```
# отправим запрос на подтверждение результата
connection.execute('''
CREATE OR REPLACE VIEW ords AS
    SELECT 
        orders.orderNumber 'Номер заказа', 
        products.productName 'Название продукта',
        products.productCode 'Код продукта',
        quantityOrdered 'Количество заказов',
        priceEach 'Цена за один',
        products.buyPrice 'Цена покупки',
        SUM(products.buyPrice * quantityOrdered) 'Итоговая сумма',
        SUM((priceEach - products.buyPrice) * quantityOrdered) 'Прибыль',
        customers.customerName 'Заказчик',
        CONCAT(customers.contactLastName, ' ', customers.contactFirstName) 'Имя представителя', 
        customers.addressLine1 'адрес'
    FROM
        orderdetails
    INNER JOIN orders
    ON orderdetails.orderNumber = orders.orderNumber
    INNER JOIN products
    ON products.productCode = orderdetails.productCode
    LEFT JOIN customers
    ON customers.customerNumber = orders.customerNumber
    GROUP by orders.orderNumber, products.productName, products.productCode
    ORDER by orders.orderNumber, products.productCode
''')
```

```
pd.read_sql('SELECT * FROM ords LIMIT 10;', connection)
```

Из полученного представления найти заказчиков и принесенную прибыль, (отсортировать по убыванию суммарной прибыли по всем)

```
pd.read_sql('''SELECT `Заказчик`, SUM(`Прибыль`) `Суммарная прибыль`
FROM ords
where `Номер заказа` in (Select orderNumber from orders where status = 'Shipped')
GROUP by `Заказчик`
Order by `Суммарная прибыль` DESC LIMIT 10;''', connection)
```
